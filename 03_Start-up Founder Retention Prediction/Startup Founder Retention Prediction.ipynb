````python
import pandas as pd
test_df = pd.read_csv('test(2).csv')
test_df.head()
````

````python
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Replace this with your target + ID column names
TARGET = 'retention_status'        # change as needed
IDCOL  = 'founder_id'            # change as needed

# Split X and y
y = train_df[TARGET]
X = train_df.drop(columns=[TARGET, IDCOL], errors='ignore')

# Test features
test = test_df.drop(columns=[IDCOL], errors='ignore')

# Detect feature types
cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()

print("Numeric cols:", len(num_cols), "Categorical cols:", len(cat_cols))
````

````python
# Numeric pipeline: median impute + scaling
num_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Categorical pipeline: most-frequent impute + OHE (dense)
cat_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Full preprocessing
pre_keras = ColumnTransformer(
    transformers=[
        ('num', num_pipe, num_cols),
        ('cat', cat_pipe, cat_cols)
    ],
    remainder='drop'
)

# Fit on full training X
pre_keras.fit(X)
````

````python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.layers import BatchNormalization, Dropout, Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
# Transform the target variable 'y' to numerical labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Apply the preprocessing to the training and test data
X_all = pre_keras.transform(X)
X_test_all = pre_keras.transform(test)

# Define the MLP model
model = Sequential([
    Dense(256, activation='gelu', input_shape=(X_all.shape[1],)),
    Dense(128, activation='relu'),
    Dense(64, activation='gelu'),

    Dense(32, activation='relu'),


    Dense(1, activation='sigmoid')
])

callback = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)



# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Print model summary
print("\n--- MLP Model Summary ---")
model.summary()

# Train the model
print("\n--- Training MLP Model ---")
history = model.fit(
        X_all, y_encoded,
        epochs=20,
        batch_size=32,
        validation_split=0.2,
        callbacks=[callback],
        verbose=1
)


print("\nMLP model training complete.")
````

## Perform Cross-Validation

### Subtask:
Implement k-fold cross-validation to get a more reliable estimate of the model's performance and macro score, reducing the bias of a single train-validation split.


**Reasoning**:
I need to implement k-fold cross-validation to evaluate the model's performance more robustly. This involves importing necessary classes, defining folds, iterating through them, cloning and training the model for each fold, and collecting and averaging the evaluation metrics.



````python
import tensorflow as tf
from tensorflow.keras.models import Sequential, clone_model
from tensorflow.keras.layers import Dense, Input, Dropout
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import KFold
from sklearn.metrics import classification_report
import numpy as np

# Assuming X_all and y_encoded are already defined from previous steps
# and le (LabelEncoder) is fitted

# 1. Define the number of folds for cross-validation
n_splits = 5

# 2. Initialize empty lists to store macro F1-scores, precision, and recall for each fold
macro_f1_scores = []
macro_precisions = []
macro_recalls = []

# 3. Initialize the KFold object
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

print(f"\n--- Performing {n_splits}-Fold Cross-Validation ---")

# Create the base model architecture (from the 'Adjust Model Architecture' step)
# This will be cloned for each fold
def create_model():
    model = Sequential([
        Input(shape=(X_all.shape[1],)),
        Dense(256, activation='relu'),
        Dropout(0.3),
        Dense(128, activation='relu'),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid') # Binary classification output
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Use the previous epochs and batch_size values
epochs_new = 50
batch_size_new = 64

# 4. Loop through each fold
for fold, (train_index, val_index) in enumerate(kf.split(X_all, y_encoded)):
    print(f"\n--- Fold {fold+1}/{n_splits} ---")

    # 5. Split X_all and y_encoded into training and validation sets for the current fold
    X_train_fold, X_val_fold = X_all[train_index], X_all[val_index]
    y_train_fold, y_val_fold = y_encoded[train_index], y_encoded[val_index]

    # 6. Create a fresh instance of the model architecture by cloning the base model
    model_fold = create_model()

    # 7. Train the model on the training data for the current fold
    history_fold = model_fold.fit(X_train_fold, y_train_fold,
                                  epochs=epochs_new,
                                  batch_size=batch_size_new,
                                  validation_data=(X_val_fold, y_val_fold),
                                  verbose=0)

    # 8. Make predictions on the validation data for the current fold
    y_val_pred_proba_fold = model_fold.predict(X_val_fold, verbose=0)
    y_val_pred_fold = (y_val_pred_proba_fold > 0.5).astype(int)

    # 9. Inverse transform the true y_val and predicted y_val_pred labels
    y_val_decoded_fold = le.inverse_transform(y_val_fold)
    y_val_pred_decoded_fold = le.inverse_transform(y_val_pred_fold.flatten())

    # 10. Calculate the classification report for the current fold
    report_fold = classification_report(y_val_decoded_fold, y_val_pred_decoded_fold, output_dict=True)

    # 11. Extract and append the 'macro avg' metrics
    macro_precisions.append(report_fold['macro avg']['precision'])
    macro_recalls.append(report_fold['macro avg']['recall'])
    macro_f1_scores.append(report_fold['macro avg']['f1-score'])

    print(f"Fold {fold+1} Macro Avg Precision: {report_fold['macro avg']['precision']:.2f}")
    print(f"Fold {fold+1} Macro Avg Recall:    {report_fold['macro avg']['recall']:.2f}")
    print(f"Fold {fold+1} Macro Avg F1-score:  {report_fold['macro avg']['f1-score']:.2f}")

# 12. Calculate the mean of the collected macro F1-scores, precision, and recall
mean_macro_precision = np.mean(macro_precisions)
mean_macro_recall = np.mean(macro_recalls)
mean_macro_f1_score = np.mean(macro_f1_scores)

# 13. Print the average macro F1-score, precision, and recall across all folds
print("\n--- Average Metrics Across All Folds ---")
print(f"Average Macro Avg Precision: {mean_macro_precision:.2f}")
print(f"Average Macro Avg Recall:    {mean_macro_recall:.2f}")
print(f"Average Macro Avg F1-score:  {mean_macro_f1_score:.2f}")
````

This involves defining a `build_model` function, setting up the `Hyperband` tuner, executing the search, and then retrieving and printing the best hyperparameters found to identify the optimal model configuration.



````python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
import keras_tuner as kt

# Ensure X_all and y_encoded are available from previous steps
# X_all: preprocessed features for training
# y_encoded: encoded target variable

# 1. Define the build_model function for Keras Tuner
def build_model(hp):
    model = Sequential()
    model.add(Input(shape=(X_all.shape[1],)))

    # Tune the number of hidden layers
    num_hidden_layers = hp.Int('num_hidden_layers', min_value=1, max_value=3, step=1)

    for i in range(num_hidden_layers):
        # Tune units in Dense layers
        model.add(Dense(units=hp.Int(f'units_{i}', min_value=32, max_value=256, step=32),
                        activation=hp.Choice(f'activation_{i}', ['relu', 'tanh', 'sigmoid'])))
        # Tune Dropout rate
        if hp.Boolean(f'dropout_{i}_enabled', default=False):
             model.add(Dropout(hp.Float(f'dropout_{i}', min_value=0.1, max_value=0.5, step=0.1)))

    # Output layer
    model.add(Dense(1, activation='sigmoid'))

    # Tune learning rate for Adam optimizer
    hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])

    model.compile(optimizer=Adam(learning_rate=hp_learning_rate),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

print("--- Starting Keras Tuner Hyperband Search ---")

# 2. Initialize the Keras Tuner Hyperband tuner
tuner = kt.Hyperband(
    build_model,
    objective='val_accuracy',
    max_epochs=50,
    factor=3,
    directory='my_dir',
    project_name='mlp_tuning',
    overwrite=True # Set to True to re-run the search from scratch
)

# Prepare data for tuning (using the full preprocessed training data)
# The validation_split argument in tuner.search will handle the split for each trial
X_tune, y_tune = X_all, y_encoded

# 3. Call tuner.search() to start the hyperparameter search
tuner.search(
    X_tune, y_tune,
    epochs=50, # This defines the maximum epochs for a trial, actual epochs will be controlled by Hyperband
    validation_split=0.2,
    callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)] # Optional: add early stopping to trials
)

print("--- Keras Tuner Hyperband Search Complete ---")

# 4. Retrieve the best hyperparameters found by the tuner
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

# 5. Print the summary of the best hyperparameters
print("\n--- Best Hyperparameters Found ---")
print(f"Number of hidden layers: {best_hps.get('num_hidden_layers')}")
for i in range(best_hps.get('num_hidden_layers')):
    print(f"  Layer {i+1} Units: {best_hps.get(f'units_{i}')}")
    print(f"  Layer {i+1} Activation: {best_hps.get(f'activation_{i}')}")
    if best_hps.Boolean(f'dropout_{i}_enabled'):
        print(f"  Layer {i+1} Dropout Rate: {best_hps.get(f'dropout_{i}')}")
    else:
        print(f"  Layer {i+1} Dropout Enabled: False")
print(f"Learning Rate: {best_hps.get('learning_rate')}")

# Retrieve the best model (optional, but useful for evaluation)
best_model = tuner.get_best_models(num_models=1)[0]

# Print a summary of the best model
print("\n--- Best Model Summary ---")
best_model.summary()
````
